```python
data=data.drop('caused_by',axis=1)
```

```python
data = data[data['incident_state'].map(lambda x: str(x)!="Active" and str(x)!="New" and str(x)!='Awaiting User Info' and str(x)!='Awaiting Vendor' and str(x)!='Awaiting Problem' and str(x)!='Awaiting Evidence' and str(x)!='-100')]
```

```python
data=data[data['active'].map(lambda x: str(x)!="True" )]
data=data.reset_index()
data.head().T
```

```python
data['closed_at']  = pd.to_datetime(data['closed_at'], format='%d/%m/%Y %H:%M')
data['resolved_at']= pd.to_datetime(data['resolved_at'], format='%d/%m/%Y %H:%M')
data['opened_at']  = pd.to_datetime(data['opened_at'], format='%d/%m/%Y %H:%M')
data['sys_created_at']= pd.to_datetime(data['sys_created_at'], format='%d/%m/%Y %H:%M')
data['sys_updated_at']  = pd.to_datetime(data['sys_updated_at'], format='%d/%m/%Y %H:%M')
data['time_before_completion']=data['resolved_at'].sub(data['opened_at'], axis=0).dt.days
```

```python
data=data.drop('closed_at',axis=1)
data=data.drop('resolved_at',axis=1)
data=data.drop('sys_created_at',axis=1)
data=data.drop('sys_updated_at',axis=1)

# On supprime ces quatres colonnes afin de ne pas donner la réponse à l'algorithme

data["opened_at_year"       ]=data["opened_at"].dt.year
data["opened_at_month"      ]=data["opened_at"].dt.month
data["opened_at_day"        ]=data["opened_at"].dt.day
data["opened_at_hour"       ]=data["opened_at"].dt.hour
data["opened_at_dayofyear"  ]=data["opened_at"].dt.dayofyear
data["opened_at_week"       ]=data["opened_at"].dt.week
data["opened_at_weekofyear" ]=data["opened_at"].dt.weekofyear
data["opened_at_dayofweek"  ]=data["opened_at"].dt.dayofweek
data["opened_at_weekday"    ]=data["opened_at"].dt.weekday
data["opened_at_quarter"    ]=data["opened_at"].dt.quarter
data=data.drop('opened_at',axis=1)
```

```python
data.active                  = data.active.astype(int)
data.knowledge               = data.knowledge.astype(int)
data.u_priority_confirmation = data.u_priority_confirmation.astype(int)
data.made_sla                = data.made_sla.astype(int)
```

```python
data['u_symptom'] = data['u_symptom'].cat.add_categories("Non_renseigner_u_symptom").fillna("Non_renseigner_u_symptom")
data['sys_created_by'] = data['sys_created_by'].cat.add_categories("Non_renseigner_sys_created_by").fillna("Non_renseigner_sys_created_by")
data['subcategory'] = data['subcategory'].cat.add_categories("Non_renseigner_subcategory").fillna("Non_renseigner_subcategory")
data['resolved_by'] = data['resolved_by'].cat.add_categories("Non_renseigner_resolved_by").fillna("Non_renseigner_resolved_by")
data['opened_by'] = data['opened_by'].cat.add_categories("Non_renseigner_opened_by").fillna("Non_renseigner_opened_by")
data['location'] = data['location'].cat.add_categories("Non_renseigner_location").fillna("Non_renseigner_location")
data['closed_code'] = data['closed_code'].cat.add_categories("Non_renseigner_closed_code").fillna("Non_renseigner_closed_code")
data['category'] = data['category'].cat.add_categories("Non_renseigner_category").fillna("Non_renseigner_category")
data['caller_id'] = data['caller_id'].cat.add_categories("Non_renseigner_caller_id").fillna("Non_renseigner_caller_id")
data['assignment_group'] = data['assignment_group'].cat.add_categories("Non_renseigner_assignment_group").fillna("Non_renseigner_assignment_group")
data['assigned_to'] = data['assigned_to'].cat.add_categories("Non_renseigner_assigned_to").fillna("Non_renseigner_assigned_to")
data['time_before_completion'] = data['time_before_completion'].fillna(data['time_before_completion'].median())
```
